```python
df= df.drop_duplicates()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Sales'], bins=30 ,kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Category', y='Sales', data=df)
plt.title('Sales by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Category', y='Profit', data=df)
plt.title('Profit by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Sales', data=df)
plt.title('Sales by Region')
plt.show()
```

```python
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
top_states.plot(kind='bar')
plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Total Sales')

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
top_products = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
top_products.plot(kind='bar')
plt.title('Top 10 Products by Sales')   
plt.show()
```

Business Insights

1. Technology category generated the highest sales.

2. Some states contribute significantly more revenue than others.

3. High discounts may reduce profit.

4. A few products generate most of the revenue.

5. Sales are concentrated in specific regions.

Recommendations

1. Increase marketing for high-performing products.

2. Reduce discounts on low-profit products.

3. Focus more on high-sales regions.

4. Improve inventory management for top-selling products.

5. Promote low-performing categories with targeted campaigns.
